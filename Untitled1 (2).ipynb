```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, to_date
import pandas as pd
from google.colab import files


# Supprimer les doublons
athletes_df = athletes_df.dropDuplicates()
medals_df = medals_df.dropDuplicates()
hosts_df = hosts_df.dropDuplicates()
results_df = results_df.dropDuplicates()

# Gérer les valeurs manquantes
athletes_df = athletes_df.dropna()
medals_df = medals_df.dropna()
hosts_df = hosts_df.dropna()
results_df = results_df.dropna()

# Uniformiser les noms de pays
def unify_country_names(df):
    df = df.withColumn('country_name', when(col('country_name') == 'Soviet Union', 'Russia')
                                          .when(col('country_name') == 'Russian Federation', 'Russia')
                                          .when(col('country_name') == 'USA', 'United States')
                                          .when(col('country_name') == 'UK', 'United Kingdom')
                                          .when(col('country_name') == 'Great Britain', 'United Kingdom')
                                          .otherwise(col('country_name')))
    return df

medals_df = unify_country_names(medals_df)
results_df = unify_country_names(results_df)

# Convertir les colonnes de date en format approprié
hosts_df = hosts_df.withColumn('game_start_date', to_date(col('game_start_date')))
hosts_df = hosts_df.withColumn('game_end_date', to_date(col('game_end_date')))

# Fusionner les DataFrames
data = medals_df.join(hosts_df, medals_df.slug_game == hosts_df.game_slug)
data = data.join(athletes_df, on='athlete_url', how='left')

# Sélectionner les caractéristiques importantes
features = data.select('game_year', 'game_season', 'event_title', 'participant_type', 'country_name', 'medal_type')

# Convertir en pandas DataFrame pour sklearn
features = features.toPandas()

# Encoder les colonnes catégorielles
features = pd.get_dummies(features, columns=['game_season', 'event_title', 'participant_type', 'country_name'])

# Créer les labels (médailles)
labels = pd.get_dummies(features['medal_type'])
features = features.drop(columns=['medal_type'])

# Diviser les données en ensembles d'entraînement et de test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Normaliser les données
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Enregistrer les DataFrames nettoyés en CSV
athletes_df.toPandas().to_csv('cleaned_olympic_athletes.csv', index=False)
medals_df.toPandas().to_csv('cleaned_olympic_medals.csv', index=False)
hosts_df.toPandas().to_csv('cleaned_olympic_hosts.csv', index=False)
results_df.toPandas().to_csv('cleaned_olympic_results.csv', index=False)

# Enregistrer le DataFrame combiné en CSV
combined_df = data.toPandas()
combined_df.to_csv('combined_olympic_data.csv', index=False)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.base import BaseEstimator, ClassifierMixin

class CustomKerasClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, neurons=1, dropout_rate=0.0, optimizer='adam', epochs=50, batch_size=32, verbose=0):
        self.neurons = neurons
        self.dropout_rate = dropout_rate
        self.optimizer = optimizer
        self.epochs = epochs
        self.batch_size = batch_size
        self.verbose = verbose
        self.model = None

    def build_model(self):
        model = Sequential()
        model.add(Dense(self.neurons, input_dim=self.input_shape, activation='relu'))
        model.add(Dropout(self.dropout_rate))
        model.add(Dense(self.output_shape, activation='softmax'))
        model.compile(loss='categorical_crossentropy', optimizer=self.optimizer, metrics=['accuracy'])
        return model

    def fit(self, X, y):
        self.input_shape = X.shape[1]
        self.output_shape = y.shape[1]
        self.model = self.build_model()
        self.model.fit(X, y, epochs=self.epochs, batch_size=self.batch_size, verbose=self.verbose)
        return self

    def predict(self, X):
        return self.model.predict(X)

    def predict_proba(self, X):
        return self.model.predict_proba(X)

    def score(self, X, y):
        _, accuracy = self.model.evaluate(X, y, verbose=0)
        return accuracy
```

```python
from sklearn.model_selection import GridSearchCV


# Fusionner les DataFrames
data = medals_df.join(hosts_df, medals_df.slug_game == hosts_df.game_slug)
data = data.join(athletes_df, on='athlete_url', how='left')

# Sélectionner les caractéristiques importantes
features = data.select('game_year', 'game_season', 'event_title', 'participant_type', 'country_name', 'medal_type')

# Convertir en pandas DataFrame pour sklearn
features = features.toPandas()

# Encoder les colonnes catégorielles
features = pd.get_dummies(features, columns=['game_season', 'event_title', 'participant_type', 'country_name'])

# Créer les labels (médailles)
labels = pd.get_dummies(features['medal_type'])
features = features.drop(columns=['medal_type'])

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Normaliser les données
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Créer le modèle KerasClassifier personnalisé
model = CustomKerasClassifier(epochs=50, batch_size=32, verbose=0)

# Définir la grille de paramètres
param_grid = {
    'neurons': [64, 128, 256],
    'dropout_rate': [0.2, 0.4, 0.6],
    'optimizer': ['adam', 'rmsprop']
}

# Rechercher les meilleurs hyperparamètres sans multiprocessing
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=1, cv=3)
grid_result = grid.fit(X_train, y_train)

# Résultats de la recherche d'hyperparamètres
print(f'Best: {grid_result.best_score_} using {grid_result.best_params_}')

```
